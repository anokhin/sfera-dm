```python
DATA_DIR = "/Users/nikolayanokhin/Desktop/Lecture-04/"
dataset = sd.fetch_olivetti_faces(data_home=DATA_DIR)
```

```python
def plot_dataset(dataset, x, y, margin=50, frame=2000., imsize=64):
    x1 = (x - x.min()) * (frame / (x.max() - x.min())) + margin
    y1 = (y - y.min()) * (frame / (y.max() - y.min())) + margin
    image = np.zeros((x1.max() + margin + imsize, y1.max() + margin + imsize)) + 255
    print "Image size is %dx%d" % image.shape
    for k in xrange(dataset.images.shape[0]):
        image[x1[k]:x1[k]+64, y1[k]:y1[k]+64] = dataset.images[k] * 255
    pl.figure(figsize=(20, 20))
    pl.imshow(image)
    pl.axis("off")
    pl.show()
```

Нарисуем сначала картинки в случайных местах

```python
x_rnd, y_rnd = np.random.randint(0, 2000, len(dataset.target)), np.random.randint(0, 2000, len(dataset.target))
plot_dataset(dataset, x_rnd, y_rnd)
```

Далее реализуем и применим MDS. Требуется реализовать MDS, используя одну из функций из пакета `scipy.optimize`: `scipy.optimize.fmin_cg`. Для того, чтобы запустить алгоритм оптимизации, потребуется реализовать критерий (`J`) и его градиент (`dJ`), и подать их в качестве параметров в алгоритм.

```python
n = len(dataset.target)
# Distances between x points
delta = ssd.squareform(ssd.pdist(dataset.data, "euclidean"))

# Starting point for optimization
y0 = np.random.random(2 * n) * 100

# Auxiliary mask can be used in J
ind_mat = np.arange(n).repeat(n).reshape((n, n))
mask = ind_mat < ind_mat.T

def J(y, *args):
    delta, mask = args
    n = delta.shape[0]
    z = y.reshape(n, 2)
    d = ssd.squareform(ssd.pdist(z, "euclidean"))
    m = (d - delta) * (d - delta) / delta / delta
    return m[mask].sum()

def dJ(y, *args):
    delta, mask = args
    n = delta.shape[0]
    z = y.reshape(n, 2)
    d = ssd.squareform(ssd.pdist(z, "euclidean"))
    grad = np.zeros(z.shape)
    for k in xrange(n):
        for j in xrange(n):
            if k != j:
                grad[k] += (d[k, j] - delta[k, j]) / delta[k, j] / delta[k, j] / d[k, j] * (z[k] - z[j])
    return 2 * grad.flatten()

args = (delta, mask)
res = so.fmin_cg(J, y0, fprime=dJ, args=args, full_output=True)
x_mds, y_mds = res[0][:n], res[0][n:]
x_mds[np.isnan(x_mds)] = 0
y_mds[np.isnan(y_mds)] = 0

plot_dataset(dataset, x_mds, y_mds)
```

Применим t-SNE. Для этого просто используем готовую реализацию из sklearn. Как результат зависит от perplexity? (Проверьте несколько значений в диапазоне 5-500)

```python
tsne = sm.TSNE(n_components=2, perplexity=30.0, verbose=1, n_iter=1000)
z = tsne.fit_transform(np.asfarray(dataset.data, dtype='float'))
z = z / z.max() * 1500
plot_dataset(dataset, z[:, 0], z[:, 1])
```
